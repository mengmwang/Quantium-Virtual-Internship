```python
# before trial, get the metric for finding control stores
df_month = df1.groupby(['STORE_NBR','MONTH']).agg({'TOT_SALES':['count','sum'],'LYLTY_CARD_NBR': 'nunique'}).reset_index()
df_month
```

```python
# correlation and distance metrics
ref = 77
df_month_ref = df_month.loc[df_month['STORE_NBR']==ref].set_index('MONTH', drop=True)
i = 1
df_month_i = df_month.loc[df_month['STORE_NBR']==i].set_index('MONTH', drop=True)
corr_ntran = df_month_ref.corrwith(df_month_i)[1]
corr_total = df_month_ref.corrwith(df_month_i)[2]
corr_ncust = df_month_ref.corrwith(df_month_i)[3]
print(corr_ntran,corr_total,corr_ncust)

distance = np.linalg.norm(df_month_ref-df_month_i)
print(distance)
```

```python
# find similar stores to store 77
ref = 77
ntran = []
total = []
ncust = []
dist = []
ref_list = []
str_list = []
for i in range(df_month['STORE_NBR'].nunique()):
    store = df_month['STORE_NBR'].unique()[i]
    df_month_ref = df_month.loc[df_month['STORE_NBR']==ref].set_index('MONTH', drop=True)
    df_month_i = df_month.loc[df_month['STORE_NBR']==store].set_index('MONTH', drop=True)
    corr_ntran = df_month_ref.corrwith(df_month_i)[1]
    corr_total = df_month_ref.corrwith(df_month_i)[2]
    corr_ncust = df_month_ref.corrwith(df_month_i)[3]
    distance = np.linalg.norm(df_month_ref-df_month_i)
    ref_list.append(ref)
    str_list.append(store)
    ntran.append(corr_ntran)
    total.append(corr_total)
    ncust.append(corr_ncust)
    dist.append(distance)
    #print(ntran,total,ncust,dist)
df_corr_77 = pd.DataFrame({'ref':ref_list, 'store':str_list, 'corr_ntran':ntran, 'corr_total':total, 'corr_ncust':ncust, 'dist':dist})
df_corr_77
```

```python
# average correlation
df_corr_77['corr_avg'] = df_corr_77[['corr_ntran','corr_total','corr_ncust']].mean(axis=1)
```

```python
# sort correlation & distance
df_corr_77.sort_values(by='corr_avg', ascending=False).head(10)
```

```python
# try statistical testing
a = df.loc[df['STORE_NBR']==ref].groupby('MONTH')['TOT_SALES'].sum()[7:10].values.tolist()
b = df.loc[df['STORE_NBR']==ctr].groupby('MONTH')['TOT_SALES'].sum()[7:10].values.tolist()
from scipy import stats
ttest=stats.ttest_ind(a,b)
alpha=0.05
if(ttest[1]<alpha):
    print('The trial store is significantly different from the control store')
else:
    print('Not significant')
# the results is not statistically significant, due to insufficient number of samples (3 month)
```
